```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

#generate x and y
x = np.linspace(0,1,101)
y = 1+x+x*np.random.random(len(x))

#assemble matrix A
#vstack stacks arrays row wise.
A = np.vstack([x,np.ones(len(x))]).T
#print("A: "+ str(A))

#turn Y into a column vector
y = y[:,np.newaxis]

#direct least squares regression 
alpha = np.dot((np.dot(np.linalg.inv(np.dot(A.T,A)),A.T)),y)
print(alpha)

plt.figure(figsize = (10,8))
#plot data points
plt.plot(x,y,'b.')
#plot line of best fit
plt.plot(x,alpha[0]*x + alpha[1],'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Using the Pseudoinverse method

```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

#generate x and y
x = np.linspace(0,1,101)
y = 1+x+x*np.random.random(len(x))

#assemble matrix A
#vstack stacks arrays row wise.
A = np.vstack([x,np.ones(len(x))]).T
#print("A: "+ str(A))

#turn Y into a column vector
y = y[:,np.newaxis]

#using the pseudoinverse method
pinv = np.linalg.pinv(A)
alpha = pinv.dot(y)
print(alpha)
```

Using numpy.linalg.lstsq

```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

#generate x and y
x = np.linspace(0,1,101)
y = 1+x+x*np.random.random(len(x))

#assemble matrix A
#vstack stacks arrays row wise.
A = np.vstack([x,np.ones(len(x))]).T
#print("A: "+ str(A))

#turn Y into a column vector
y = y[:,np.newaxis]

alpha = np.linalg.lstsq(A,y,rcond = None)[0]
print(alpha)
```
